```python
import pandas as pd
from sklearn import linear_model
import  matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/apratim777/apratim777/master/Position_Salaries.csv"
df= pd.read_csv(url)
print(df.head())
```

```python
#choose x and y
# x is Level column
x=df.iloc[:,1:2].values
y=df.iloc[:,2].values
print(x.shape)
print(y.shape)
```

```python
reg=linear_model.LinearRegression()
reg.fit(x,y)
acc=reg.score(x,y)
print(acc)
```

```python
plt.xlabel('Level')
plt.ylabel('Salary')
plt.scatter (df.Level,df.Salary)
plt.plot(df.Level,reg.predict(x),color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures
reg2=PolynomialFeatures(degree=5)
x_poly=reg2.fit_transform(x)
print(x_poly.shape)
```

```python
reg3=linear_model.LinearRegression()
reg3.fit(x_poly,y) 
acc=reg3.score(x_poly,y) 
print(acc)
```

```python
plt.xlabel('Level')
plt.ylabel('Salary')
plt.scatter (df.Level,df.Salary)
plt.plot(df.Level,reg3.predict(reg2.fit_transform(x) ),color='red')
```
