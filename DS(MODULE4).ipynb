```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats# or we can write import scipy.stats as st
url="https://raw.githubusercontent.com/p-p-p-p/ml/main/info3.csv"
url1="https://raw.githubusercontent.com/p-p-p-p/ml/main/info4.csv"
health_information3 = pd.read_csv(url, header=0, sep=",")

x = health_information3["Average_Pulse"]
y = health_information3 ["Calorie_Burnage"]

slope, intercept, r, p,  std_err= stats.linregress(x, y)# or slope, intercept, r, p,  std_err= st.linregress(x, y)
print("slope=",slope)
print("intercept=",intercept)
print("Rvalue=",r)
print("Pvalue=",p)
print("standard error=",std_err)

def myfunc(x):
 return slope * x + intercept

new_y = list(map(myfunc, x))
print(new_y)

plt.scatter(x, y)
plt.plot(x, new_y,color="red")#or we can write plt.plot(x, slope * x + intercept)
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel ("Calorie_Burnage")
plt.show()
```
