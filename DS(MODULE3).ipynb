```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

health_information3= pd.read_csv(url, header=0, sep=",")

health_information3_heatmap = health_information3.corr()

axis_corr=sns.heatmap(health_information3_heatmap,vmin=-1, vmax=1, center=0,cmap=sns.diverging_palette(10, 300,  n=500),
square=True)
print(axis_corr)

plt.show()
```

```python
#BEACH EXAMPLE
import pandas as pd
import matplotlib.pyplot as plt


drowning = {"drowning_accident": [20,40,60,80,100,120,140,160,180,200],
"ice_cream_sale": [20,40,60,80,100,120,140,160,180,200]}
drowning = pd.DataFrame(data=drowning)
print(drowning)
drowning.plot(x="ice_cream_sale", y="drowning_accident", kind="scatter")
plt.show()

correlation_beach = drowning.corr()
print(correlation_beach)
```
