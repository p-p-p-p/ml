```python
import pandas as pd
url="https://raw.githubusercontent.com/p-p-p-p/ml/main/info2.csv"
health_information = pd.read_csv(url, header=0, sep=",")

print(health_information)
```

```python
import matplotlib.pyplot as plt

health_information.plot(x ='Average_Pulse', y='Calorie_Burnage', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()
```

```python
def slope(x1, y1, x2, y2):
  s = (y2-y1)/(x2-x1)
  return s

print (slope(80,240,90,260))
```
